```python
import pandas as pd
from pathlib import Path
```

```python
# import countryContinent.csv
country_continent_data = Path("Resources/countryContinent.csv")

country_df = pd.read_csv(country_continent_data, encoding='latin1')
country_df.head()
```

```python
# import Global_Dataset_of_Inflation.csv
inflation_data = Path("Resources/Global_Dataset_of_Inflation.csv")

inflation_df = pd.read_csv(inflation_data, encoding='latin1')
inflation_df.head()
```

```python
# show columns in inflation_df
inflation_df.columns
```

```python
# pull desired columns from original dataframe
inflation_df = inflation_df[['Country Code', 'Series Name', '2002', '2003',
       '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021',
       '2022']]
inflation_df.head()
```

```python
# show columns in country_df
country_df.columns
```

```python
# pull desired columns from original dataframe
country_df = country_df[['code_3', 'country', 'continent', 'sub_region']]
country_df.head()
```

```python
# rename coumsn in country_df to match with inflation_df
country_df = country_df.rename(columns={'code_3': 'Country Code', 'country':'Country', 'continent':'Continent', 'sub_region':'Subregion'})
country_df.head()
```

```python
# check for any missing data in country_df
country_df.count()
```

```python
# drop any rows with missing values
country_df = country_df.dropna(how='any')
```

```python
# check to make sure all columns are now equal
country_df.count()
```

```python
# export country_df as csv without index
country_df.to_csv('Output/countries.csv', index=False, header=True)
```

```python
# check inflation type
inflation_df['Series Name'].unique()
```

```python
# create DataFrame for 'Headline Consumer Price Inflation'
headline_inflation = inflation_df[inflation_df['Series Name'] == 'Headline Consumer Price Inflation']

# drop 'Series Name' column
headline_inflation = headline_inflation.drop('Series Name', axis='columns')

# drop any rows with missing values
headline_inflation = headline_inflation.dropna(how='any')

headline_inflation.head(10)
```

```python
# merge headline_inflation with country_df
headline_inflation = pd.merge(country_df, headline_inflation, on='Country Code', how='right')

headline_inflation.head(10)
```

```python
# create DataFrame for 'Energy Consumer Price Inflation'
energy_inflation = inflation_df[inflation_df['Series Name'] == 'Energy Consumer Price Inflation']

# drop 'Series Name' column
energy_inflation = energy_inflation.drop('Series Name', axis='columns')

# drop any rows with missing values
energy_inflation = energy_inflation.dropna(how='any')

# reset index
energy_inflation = energy_inflation.reset_index(drop=True)

energy_inflation.head(10)
```

```python
# create DataFrame for 'Food Consumer Price Inflation'
food_inflation = inflation_df[inflation_df['Series Name'] == 'Food Consumer Price Inflation']

# drop 'Series Name' column
food_inflation = food_inflation.drop('Series Name', axis='columns')

# drop any rows with missing values
food_inflation = food_inflation.dropna(how='any')

# reset index
food_inflation = food_inflation.reset_index(drop=True)

food_inflation.head(10)
```

```python
# create DataFrame for 'Official Core Consumer Price Inflation'
consumer_inflation = inflation_df[inflation_df['Series Name'] == 'Official Core Consumer Price Inflation']

# drop 'Series Name' column
consumer_inflation = consumer_inflation.drop('Series Name', axis='columns')

# drop any rows with missing values
consumer_inflation = consumer_inflation.dropna(how='any')

# reset index
consumer_inflation = consumer_inflation.reset_index(drop=True)

consumer_inflation.head(10)
```

```python
# create DataFrame for 'Producer Price Inflation'
producer_inflation = inflation_df[inflation_df['Series Name'] == 'Producer Price Inflation']

# drop 'Series Name' column
producer_inflation = producer_inflation.drop('Series Name', axis='columns')

# drop any rows with missing values
producer_inflation = producer_inflation.dropna(how='any')

# reset index
producer_inflation = producer_inflation.reset_index(drop=True)

producer_inflation.head(10)
```

```python
# merge country_df into inflation_df on 'Country Code'
merged_df = pd.merge(country_df, inflation_df, on='Country Code', how='right')
merged_df
```

```python
# export merged_df as csv without index
merged_df.to_csv('Output/country_inflation_merged.csv', index=False, header=True)
```
