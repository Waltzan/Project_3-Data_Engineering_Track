```python
import pandas as pd
from pathlib import Path
```

```python
# import countryContinent.csv
country_continent_data = Path("Resources/countryContinent.csv")

country_df = pd.read_csv(country_continent_data, encoding='latin1')
country_df.head()
```

```python
# import Global_Dataset_of_Inflation.csv
inflation_data = Path("Resources/Global_Dataset_of_Inflation.csv")

inflation_df = pd.read_csv(inflation_data, encoding='latin1')
inflation_df.head()
```

```python
# show columns in inflation_df
inflation_df.columns
```

```python
# pull desired columns from original dataframe
inflation_df = inflation_df[['Country Code', '2002', '2003',
       '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021',
       '2022']]
inflation_df.head()
```

```python
# show columns in country_df
country_df.columns
```

```python
# pull desired columns from original dataframe
country_df = country_df[['code_3', 'country', 'continent', 'sub_region']]
country_df.head()
```

```python
# rename coumsn in country_df to match with inflation_df
country_df = country_df.rename(columns={'code_3': 'Country Code', 'country':'Country', 'continent':'Continent', 'sub_region':'Subregion'})
country_df.head()
```

```python
# check for any missing data in country_df
country_df.count()
```

```python
# drop any rows with missing values
country_df = country_df.dropna(how='any')
```

```python
# check to make sure all columns are now equal
country_df.count()
```

```python
# export country_df as csv without index
country_df.to_csv('Output/countries.csv', index=False, header=True)
```

```python
# check for any missing values in inflation_df
inflation_df.count()
```

```python
# drop any rows with missing values
inflation_df = inflation_df.dropna(how='any')
```

```python
# check to make sure all columns are now equal
inflation_df.count()
```

```python
# export inflation_df as csv without index
inflation_df.to_csv('Output/inflation.csv', index=False, header=True)
```

```python
# merge country_df into inflation_df on 'Country Code'
merged_df = pd.merge(country_df, inflation_df, on='Country Code', how='right')
merged_df
```

```python
# export merged_df as csv without index
merged_df.to_csv('Output/country_inflation_merged.csv', index=False, header=True)
```
